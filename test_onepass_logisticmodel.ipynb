```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Read the loss data

input_dim = 64
filename = f"logs/training_loss_in{input_dim}_swa.log"
loss_data = pd.read_csv(filename, header='infer', dtype=float)

# Plot both losses
plt.plot(loss_data['batch'][1:][loss_data['loss'][1:] > 0], loss_data['loss'][1:][loss_data['loss'][1:] > 0], label='Average Loss', linestyle='-', color='b')
plt.plot(loss_data['batch'][1:][loss_data['swa_loss'][1:] > 0], loss_data['swa_loss'][1:][loss_data['swa_loss'][1:] > 0], label='SWA Loss', linestyle='--', color='b')
plt.xlabel('Batch')
plt.ylabel('Loss')
plt.title('Streaming Loss')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.axis('equal')

x = loss_data['batch'][1:]
plt.plot(x, 2*float(input_dim)/x * (1.0+np.log(1+4*x/float(input_dim))), color='black')
plt.show()
```
