```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read the loss data

input_dim = 10
hidden_dims_model = [50, 50]
output_dims = 2
hidden_dims_agent = [50, 50]
filename = f"training_loss_in{input_dim}_hid{hidden_dims_model[0]}_out{output_dims}_hid{hidden_dims_agent[0]}.log"
filename2 = f"training_loss_in{input_dim}_hid{hidden_dims_model[0]}_out{output_dims}_hid{100}.log"
filename3 = f"training_loss_in{input_dim}_hid{hidden_dims_model[0]}_out{output_dims}_hid{200}.log"
filename4 = f"training_loss_in{input_dim}_hid{hidden_dims_model[0]}_out{output_dims}_hid{25}.log"
loss_data = pd.read_csv(filename, header='infer', dtype=float)
loss_data2 = pd.read_csv(filename2, header='infer', dtype=float)
loss_data3 = pd.read_csv(filename3, header='infer', dtype=float)
loss_data4 = pd.read_csv(filename4, header='infer', dtype=float)

# Plot both losses
plt.plot(loss_data['epoch'], loss_data['train_loss'], label='Train Loss 50', linestyle='--', color='b')
plt.plot(loss_data['epoch'], loss_data['test_loss'], label='Test Loss 50', linestyle='-', color='b')
plt.plot(loss_data2['epoch'], loss_data2['train_loss'], label='Train Loss 100', linestyle='--', color='r')
plt.plot(loss_data2['epoch'], loss_data2['test_loss'], label='Test Loss 100', linestyle='-', color='r')
plt.plot(loss_data3['epoch'], loss_data3['train_loss'], label='Train Loss 200', linestyle='--', color='g')
plt.plot(loss_data3['epoch'], loss_data3['test_loss'], label='Test Loss 200', linestyle='-', color='g')
plt.plot(loss_data4['epoch'], loss_data4['train_loss'], label='Train Loss 25', linestyle='--', color='y')
plt.plot(loss_data4['epoch'], loss_data4['test_loss'], label='Test Loss 25', linestyle='-', color='y')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Test Loss')
plt.legend()
plt.show()
```
